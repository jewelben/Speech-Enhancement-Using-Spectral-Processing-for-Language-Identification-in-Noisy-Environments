```python

import numpy as np
import matplotlib.pyplot as plt
import librosa as lb
import librosa.display as lbd
import soundfile
import os
```

# Spectrogram

```python
def showSpectrogram(x, nFFT, title):
    stft = lb.stft(x, n_fft=nFFT, win_length= nFFT)
    stft_db = lb.amplitude_to_db(np.abs(stft),ref=np.max)
    lbd.specshow(stft_db,x_axis='time',y_axis='linear')
    plt.ylim([0, 4000])
    plt.colorbar(format="%+2.f dB")
    plt.title(title)
```

# White Noise

```python
x1, fs = lb.load(r"F:\SSP\Project\noises\noises\white.wav")
x1 = x1[5000:15000]
plt.figure(figsize=(16, 10))
plt.subplot(211)
plt.plot(x1)
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.title("White Noise")
plt.subplot(212)
showSpectrogram(x1, 512, "White Noise Spectrogram")
plt.savefig("White Noise Spectrogram.png")
```

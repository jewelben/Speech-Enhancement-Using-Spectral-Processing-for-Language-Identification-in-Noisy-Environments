```python
import numpy as np
import wave
import matplotlib.pyplot as plt
import math 
import scipy.special as sp
import librosa as lb
import librosa.display as lbd
from numpy.random import randint
```

```python
def zff ( signal , sr , duration = 0.01 )  :

        mov_avg_filter_len = int ( sr * duration )
        # signal = np.array ( signal )
        #  differentiating the signal
        signal = np.diff ( signal )

        # integrate twice
        signal = np.cumsum ( np.cumsum ( signal )  )
        kernel = np.ones ( mov_avg_filter_len ) / mov_avg_filter_len
        signal -=  np.convolve ( signal , kernel , mode = "same")
        return signal
```

```python
#load audio file
y, sr = lb.load("hello.wav", sr=None)

plt.subplot(211)
lbd.waveshow(y, sr=sr)

#apply zff
zff_signal = zff(y, sr, duration=0.01)

plt.subplot(212)
lbd.waveshow(zff_signal, sr=sr)
```
